Plotting oriented "points"
----------------------

First, we generate two random data sets of oriented points:

```python
import numpy as np

x, y = 4000 * np.random.rand(2, 500)
angle = 2*np.pi * np.random.rand(500)
size = 1 + 9 * np.random.rand(500)

u = size * np.sin(angle)
v = size * np.cos(angle)

# random split in two sets (1/3 and 2/3 of the total sample)
rnd_type = np.random.rand(500)
mask = rnd_type > 0.33
```

And now we plot them using quiver:

```python
import matplotlib.pyplot as plt

plt.quiver(x[mask], y[mask], u[mask], v[mask], pivot='middle', headwidth=0)
plt.quiver(x[~mask], y[~mask], u[~mask], v[~mask], pivot='middle', headwidth=0, color='red')
plt.show()
```
